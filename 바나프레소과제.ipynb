```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
def get_info():
    driver = webdriver.Chrome()
    driver.implicitly_wait(3)
    url = 'https://banapresso.com/store'
    driver.get(url)

    
    info_list = []
    
    for i in range(1, 12):
        try:
            button = driver.find_element(By.XPATH, f'//li/a[contains(text(), "{i}")]')
            button.click()
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, 'store_name_map')))

            elements = driver.find_elements(By.CLASS_NAME, 'store_name_map')
            info_list.extend([{
                'name': element.find_element(By.TAG_NAME, 'i').text,
                'address': element.find_element(By.TAG_NAME, 'span').text.strip()
            } for element in elements])

            if i % 5 == 0:
                next_button = driver.find_element(By.XPATH, '//a[contains(text(), "다음")]')
                next_button.click()
                WebDriverWait(driver, 10).until(EC.staleness_of(elements[0]))
        except NoSuchElementException:
            break
    driver.quit()
    return info_list

def print_info(info_list):
    for info in info_list:
        print('Name:', info['name'])
        print('Address:', info['address'])

def save_to_excel(info_list, file_name):
    df = pd.DataFrame(info_list)
    df.to_excel(file_name, index=False)

info_list = get_info()
print_info(info_list)
save_to_excel(info_list, 'store_info.xlsx')
```
